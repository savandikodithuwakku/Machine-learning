```python
import numpy as np
data = np.array([[26, 50000],
                 [29, 70000],
                 [34, 55000],
                 [31, 41000]])
```

## Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
scaled_data
```
