```python
import pandas as pd
data = pd.read_csv("student.csv")
data['result']
```

```python
result_category = data['result']    #pandas series
```

```python
from sklearn.preprocessing import LabelEncoder
obj = LabelEncoder()
result = obj.fit_transform(result_category)
```

```python
result
```

```python
from sklearn.preprocessing import LabelBinarizer
obj = LabelBinarizer()
result = obj.fit_transform(result_category)    
```

```python
result #converted to numerical values and using one hot encoding
#LabelEncoder------->LabelBinarizer
```

```python
obj.classes_        #all categories
 
```

```python
from sklearn.preprocessing import OneHotEncoder
# result = obj.fit_transform(result_category) OneHotEncoder will not accept a 1D array directly. You need to reshape it to 2D.
result_reshaped = result_category.values.reshape(-1, 1) # Convert pandas Series → NumPy 2D array
result_df = result_category.to_frame()  # method2: Convert the pandas Series to a DataFrame, which is 2D  

obj = OneHotEncoder(sparse_output=False)
# OneHotEncoder by default returns a sparse matrix Sparse matrices save memory when you have a lot of zeros (common in one-hot encoding), 
# but they don’t display like a normal array, not a normal NumPy array. You are telling it: “I don’t want a sparse matrix, give me a normal NumPy array instead.”

result_onehot = obj.fit_transform(result_reshaped)

result_onehot#converted to numerical values and using one hot encoding
```
